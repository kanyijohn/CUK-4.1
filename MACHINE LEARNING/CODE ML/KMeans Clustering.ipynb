```python
scaler=StandardScaler()
dfstd= scaler.fit_transform(dataset)
```

```python
wcss=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters= i,init='k-means++', random_state=42)
    kmeans.fit(dfstd)
    wcss.append(kmeans.inertia_)

#evaluates how well the data points are grouped within each cluster
```

```python
#plotting the 10 clusters
plt.figure(figsize=(10,8))
plt.plot(range(1,11), wcss, marker='o', linestyle='-', color='blue')
plt.xlabel('Number of clusters')
plt.ylabel('wcss')
plt.title('K-Means Clustering')
plt.show()
```

```python
df_seg=dfstd.copy()
df_seg=pd.DataFrame(data=dfstd, columns=dataset.columns)
df_seg['segmentk-means']=kmeans.labels_
```
