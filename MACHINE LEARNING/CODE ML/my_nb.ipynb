```python
sns.distplot(df_sal['Salary'])
pnt.show()
```

```python
pnt.scatter(df_sal['YearsExperience'], df_sal['Salary'], color='lightcoral')
pnt.title('Salary vs Years of Experience')
pnt.xlabel('Years of experience')
pnt.ylabel ('Salary')
```

```python
x = df_sal.iloc[: , :1]
y = df_sal.iloc[: , 1:]
X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
regressor = LinearRegression()
regressor.fit(X_train, Y_train)
```

```python
print(regressor)
```

```python
y_pread_test = regressor.predict(X_test)
y_pread_test = regressor.predict(X_train)
```

```python
pnt.scatter(X_train, Y_train, color = 'blue')
pnt.plot(X_train, y_pread_test, )
pnt.title('Salry vs Experience of training set')
pnt.xlabel('Years of experience')
pnt.ylabel('Salary')
pnt.legend(['X_train/pread_test(Y_test)', 'X_train/Y_train'], title='Sal/Exp', 
          loc='best', facecolor = 'white')
```

```python
pnt.scatter(X_test, Y_test, color = 'purple')
pnt.plot(X_train, y_pread_test, )
pnt.title('Salry vs Experience of training set')
pnt.xlabel('Years of experience')
pnt.ylabel('Salary')
pnt.legend(['X_train/pread_test(Y_test)', 'X_train/Y_train'], title='Sal/Exp', 
          loc='best', facecolor = 'white')
```

```python
print (f'Coefficients: {regressor.coef_}')
```
